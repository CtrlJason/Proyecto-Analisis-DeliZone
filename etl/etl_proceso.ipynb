### Importamos las librerias

```python
import pandas as pd
import os
```

### Inicialmente mostramos los datos de las publicaciones

```python
print("Iniciando proceso ETL de las publicaciones...")
df_origin = pd.read_csv('../data/publicaciones.csv')
df = df_origin.copy()

# Mostramos los datos de las publicaciones
df
```

### Mostramos los tipos de dato de cada columna

```python
df.info()
```

### Eliminacion de columnas que no son de utilidad

```python
COLUMNS_TO_DROP = [
    'Identificador de la cuenta',
    'Nombre de usuario de la cuenta',
    'Nombre de la cuenta', 'Duración (segundos)',
    'Comentario sobre los datos',
    'Comentarios',
    'Enlace permanente',
    'Fecha',
]
df = df.drop(COLUMNS_TO_DROP, axis=1)
df
```

Separacion de hastags de la descripción

```python
# Mostramos una de las descripciones para obtener la informacion completa
df.iloc[0]['Descripción']
```

```python
# Ahora obtenemos unicamente los numerales de todas las descripciones
df['Descripción'].str.findall('#')
# se visualizaran todos los # que esten en la descripción
```

```python
# Ahora obtenemos los Hashtags completos
df['Descripción'].str.findall(r'#\w+') # findall nos sirve para buscar partes de un string en especifico
```

```python
# Ahora creamos separamos los hastag de la descripcion
df['Hashtags'] = df['Descripción'].str.findall(r'#\w+')
df
```

### Eliminacion de los Hashtags de la descripción

```python
# Eliminar los hashtags de la columna 'Descripción'
df['Descripción'] = df['Descripción'].str.replace(r'#\w+|\n|\r', '', regex=True).str.strip()

# Verificar el resultado
df.head()
```

### Creación de la columna fecha

```python
df['Fecha de publicación'] = df['Hora de publicación'].str.split().str[0]

# Verificar el resultado
df.head()
```

### Eliminacion de la hora dentro de la fecha

```python
df['Hora de publicación'] = df['Hora de publicación'].str.split().str[1]

# Verificar el resultado
df.head()
```

### Convertimos la columna de Hashtag para mostrarlos correctamente

```python
df['Hashtags'] = df['Hashtags'].apply(lambda x: str(x).replace(",", "").replace("[", "").replace("]", "").replace("'", ""))

# Verificar el resultado
df
```

### Tratamos los datos nulos y los transofrmamos

```python
df['Seguimientos'] = df['Seguimientos'].fillna(0)

df
```

### Reduccion de texto en Tipo de publicacion

```python
df['Tipo de publicación'] = df['Tipo de publicación'].str.replace(' de Instagram', '', regex=False)
df
```

### Creamos las zonas horarios

```python
# Función para asignar el tipo de día según la hora
def obtener_zona_horaria(hora):
    if pd.isna(hora):
        return None  # O puedes asignar 'Desconocido' si prefieres
    # Extraemos la hora de la cadena de texto en 'Hora de publicación'
    hora_int = int(hora.split(':')[0])  # Esto obtiene la hora (parte antes de ':')
    
    if 0 <= hora_int < 6:
        return "Madrugada"
    elif 6 <= hora_int < 12:
        return "Mañana"
    elif 12 <= hora_int < 18:
        return "Tarde"
    else:
        return "Noche"

# Aplicamos la función a la columna 'Hora de publicación' y creamos la nueva columna 'Zona Horaria'
df['Zona Horaria'] = df['Hora de publicación'].apply(obtener_zona_horaria)

# Verificamos el resultado
df
```

### Ordenamiento de columnas

```python
columnas_ordenadas = [
    'Identificador de la publicación',
    'Fecha de publicación',
    'Hora de publicación',
    'Zona Horaria',
    'Descripción',
    'Hashtags',
    'Tipo de publicación',
    'Alcance',
    'Impresiones',
    'Me gusta',
    'Veces que se guardó',
    'Veces que se compartió',
    'Seguimientos',
    
    # Agrega las demás columnas según el orden que necesites
    # 'Columna_extra1', 'Columna_extra2', etc.
]

# Reordenamos el DataFrame con el nuevo orden de columnas
df = df[columnas_ordenadas]

# Exportamos el archivo a CSV
df.to_csv('../data/datos_trabajados/publicaciones.csv', index=False)

df
```
